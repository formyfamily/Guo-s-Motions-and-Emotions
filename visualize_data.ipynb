```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, VarianceThreshold
from sklearn.feature_selection import f_regression

dev_visual_features = np.load('data/dev_visual_features.npy')
dev_annotations = np.load('data/dev_annotations.npy')
dev_visual_features = preprocessing.minmax_scale(dev_visual_features, feature_range=(0, 1), axis=0)
print(dev_visual_features.shape)
dev_visual_features_selected = VarianceThreshold(threshold=0.01).fit_transform(dev_visual_features)
print(dev_visual_features_selected.shape)
dev_visual_features_selected_0 = SelectKBest(f_regression, k=100).fit_transform(dev_visual_features_selected, dev_annotations[:,0])
print(dev_visual_features_selected_0.shape)
dev_visual_features_selected_1 = SelectKBest(f_regression, k=100).fit_transform(dev_visual_features_selected, dev_annotations[:,1])
print(dev_visual_features_selected_1.shape)
```

```python
print(dev_annotations.shape)
dev_visual_features_selected = pd.DataFrame(dev_visual_features_selected)
dev_visual_features_selected_0 = pd.DataFrame(dev_visual_features_selected_0)
dev_visual_features_selected_1 = pd.DataFrame(dev_visual_features_selected_1)
dev_annotations = pd.DataFrame(dev_annotations)
```
